```python
text0 = '@肖战工作室 你还行不行！诗写好了没！想看帅哥 ?'
# text0 = fi.read()  # 获取文本内容
text1 = re.sub("@([\s\S]*?):", "", text0)  # 去除@ ...：
text2 = re.sub("\[([\S\s]*?)\]", "",text1)  # [...]：
text3 = re.sub("@([\s\S]*?)", "", text2)  # 去除@...
text4 = re.sub("[\.\!\/_,$%^*(+\"\']+|[+——！，。？、~@#￥%……&*（）]+", "", text3)  # 去除标点及特殊符号
text5 = re.sub("[^\u4e00-\u9fa5]", "", text4)  # 去除所有非汉字内容（英文数字）
new_text = jieba.cut(text5, cut_all=False)  # 精确模式
str_out = ' '.join(new_text)
str_out
```

```python
import re
import jieba
import numpy as np
import pandas as pd
```

```python
wb=pd.read_csv("weibo1(1).csv",encoding="gb18030",delimiter="\t")
```

```python
wb.head()
```

```python
frame = pd.DataFrame(wb)
frame = pd.DataFrame(wb, columns=['微博正文', '点赞数', '转发数', '评论数'])
frame
```

```python
# 去除“？”和“空格”等无意义符号
def sub1(x):
    x = re.sub(r"\s","", x)
    x = re.sub(r"\?*","", x)
    return x
frame['微博正文'] = frame['微博正文'].apply(sub1)
frame
```

```python
# 识别图片、链接、TAG、@偶像等内容
def contain_pic(x):
    x_mark = re.search("原图", x)
    p = re.search(r"\[组图共\d{1,2}张\]原图", x)
    if p != None:
        x_mark = p
    return x_mark
def contain_link(x):
    return re.findall(r"http\:\/\/[0-9a-zA-Z\_\.\/]+", x)
def contain_tag(x):
    tag_rule = re.compile(r"\#(.*?)\#")
    return tag_rule.findall(x)
def contain_id(x):
    return re.findall(r"\@X玖少年团肖战DAYTOY", x)
frame['附图'] = frame['微博正文'].apply(contain_pic)
frame['附链接'] = frame['微博正文'].apply(contain_link)
frame['TAG'] = frame['微博正文'].apply(contain_tag)
frame['明星'] = frame['微博正文'].apply(contain_id)
frame
```

```python
# 去“附图”“链接”等标示性文字并标注
def sub_mark(x):
    x = re.sub(r"\[组图共\d{1,2}张\]","", x)
    x = re.sub(r"原图","", x)
    x = re.sub(r"http\:\/\/[0-9a-zA-Z\_\.\/]+","", x)
    return x
def sub_tag(x):
    tag_rule = re.compile(r"(\#.*?\#)")
    return tag_rule.sub("", x)
frame['清洗正文'] = frame['微博正文'].apply(sub_mark)
frame['清洗正文'] = frame['清洗正文'].apply(sub_tag)
frame.to_csv("/users/mac/git/ghost-on-web/data_clean/output.csv")
frame
```

```python
def clean_chaohua(x):
    f1 = re.search(r"我为贡献", x)
    f2 = re.search(r"我在肖战最好的你超话等级达到", x)
    if f1 or f2 != None:
        return np.nan
    else:
        return x
frame['正文'] = frame['清洗正文'].apply(clean_chaohua)
frame = frame.dropna()
frame.to_csv("/users/mac/git/ghost-on-web/data_clean/output_clean.csv")
frame
```
